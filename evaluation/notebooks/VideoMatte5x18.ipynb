```python
import pandas as pd
```

```python
metrics_csv = "/media/andivanov/DATA/experiments/VideoMatte5x18/out/metrics.csv"
```

```python
df = pd.read_csv(metrics_csv, index_col="clipname")
df
```

```python
df[df.index.str.contains("0004_")].sort_values(by=['pha_mad'], ascending=False)
```

```python
df[df.index.str.contains("0000_")].sort_values(by=['pha_mad'], ascending=False)
```

#### Get foreground and background name

```python
df['fgr'] = df.index.map(lambda x: x.split("_", 1)[0])
df['bgr'] = df.index.map(lambda x: "".join(x.split("_", 1)[1:]))
df
```

```python
df.drop(["dynamic_mean", "semi_dynamic_mean", "static_mean"], inplace=True)
df
```

#### Average per foreground and background 

```python
df.groupby('fgr').mean().sort_values(by=["pha_mad"], ascending=False)
```

```python
df.groupby('bgr').mean().sort_values(by=["pha_mad"], ascending=False)
```
